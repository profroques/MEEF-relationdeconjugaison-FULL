```python
"""
Programme proposé par S. Roques
Académie de Lyon - Février 2024
"""

import matplotlib.pyplot as plt
import numpy as np

fp = 12.5 #distance focale en cm

#valeur de OA en mesures algébriques en cm
OA =[]
for i in range (-80,-10,5) :
    OA.append(i)
print (OA)
print("longueurs de lal iste OA valeurs expérimentales  : ", len(OA))

#Création de la liste de l'inverse de OA :
invOA=[]
for i in (OA):
    invOA.append(1/i)
print("longueur de la liste 1/OA :",len(invOA))



#Création de la liste des valeurs THEORIQUES de 1/OA' :
invOAp = []
for i in (OA):
    invOAp.append(1/fp+1/i)
print()
print("longueur de la liste 1/OA' - valeurs théoriques :",len(invOAp))

#Création de la liste des valeurs THEORIQUES de OA' :
OAp=[]
for i in (invOAp) :
    OAp.append(1/i)

print()
print("Valeurs théoriques de OA' =",OAp)
print("longueur de la liste des valeurs théoriques OA' :",len(OAp))
print()

#Création de la liste des valeurs expérimentales de OA'
# Cette liste est issue de mesures réalisées de OA' pour différentes de valeurs de OA
# La création de la liste se fait donc par saisie manuelle
OAp_expe = [14.6,15.0,15.15,15.4,15.9,16.3,16.5,17.5,18.6,19.8,21.8,25.4,33.2,74.4]
print("longueur de la liste des valeurs expérimentales OA'_expé", len(OAp_expe))


#Création de la liste de l'inverse des valeurs expérimentales 1/OA'_expe
invOAp_expe=[]
for i in (OAp_expe):
    invOAp_expe.append(1/i)

print("longueur de la liste de valeurs expérimentales 1/OA'_expé :",len(invOAp_expe))
    
    
    


# Calcul de la droite de régression linéaire
coefficients = np.polyfit(invOA, invOAp_expe, 1)
# Génération des valeurs prédites
predicted_values = np.polyval(coefficients, invOA)

# Tracé graphique
plt.figure(figsize=(8, 8))

# Premier graphique : distance lentille-image en fonction de la distance lentille-objet
plt.subplot(2, 1, 1)
plt.plot(OA, OAp, 'kx',label="OA'=f(OA)")
plt.plot(OA,OAp_expe,'r+',label = "OA' expé = f(OA)")

# Ajouts au graphique
plt.title("OA'=f(OA)")
plt.xlabel("Position de l\'objet OA barre (cm)")
plt.ylabel("Position de l\'image OA' barre (cm)")
plt.legend()
plt.grid(True)


# Deuxième graphique : Inverse de la distance (lentille-image) en fonction de l'inverse de la distance (lentille-objet)
plt.subplot(2, 1, 2)
plt.plot(invOA, invOAp_expe, "b+", label="1/OAp_expe=f(1/OA)")
plt.plot(invOA, predicted_values, label=f"Droite de régression\ny = {coefficients[0]:.2f}x + {coefficients[1]:.2f}",color = 'red',linewidth=1, linestyle='-')

plt.ylabel("1/OA' (1/cm)")
plt.xlabel("1/OA (1/cm)")

plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
print("Equation de la droite de la modélisation :\n 1/OA'=",f'{coefficients[0]:.2f}',"*1/OA + ",f'{coefficients[1]:.2f}')

#A propos de la distance focale :
fp_modelise = (1/coefficients[1]) #résultat en cm
print("Valeur de la distance focale à partir de la modélisation : ", f'{fp_modelise:.1f}', " cm")
```
